```python
#3.

# To perform this, we are going to create a new column in the "data" DataFrame,
# that is the result of the division between the "GDP" column and the "Population" column.
# Then we apply the same reasoning as the previous exercise to the new column. We finish
# the exercise by printing the final answer, where we round the value for the average
# and for the standard deviation.

data["GDP per Capita"]=data["GDP"]/data["Population"]
average_gdp_capita=data.loc[pop9_11]["GDP per Capita"].mean()
std_gdp_capita=data.loc[pop9_11]["GDP per Capita"].std()
print(f"The average GDP is {round(average_gdp_capita,2)}, while the standard deviation is {round(std_gdp_capita,2)}.")
```

```python
#4.

# We start by creating a function that receives as an argument the population of a country
# and returns as output to which class size it belongs. When the population has a NaN value
# it also returns a NaN value.

def size(pop):
    if pop<=1000000:
        return "Tiny"
    elif pop>1000000 and pop<=5000000:
        return "Very Small"
    elif pop>5000000 and pop<=15000000:
        return "Small"
    elif pop>15000000 and pop<=30000000:
        return "Medium"
    elif pop>30000000 and pop<=100000000:
        return "Large"
    elif pop>100000000:
        return "Huge"
    elif np.isnan(pop):
        return np.nan
    
# Then we are going to create a new column in the "data" DataFrame by using the .map() method
# on the column "Population" with the new "size" function as the argument.
# The .map() method is used for substituting each value in a series with another value,
# that in this case is derived from a function ("size" function, previously created).
    
data["Size"]=data["Population"].map(size)

# Then, finally we use .groupby() to combine the size classes and apply a function with .agg(),
# combining the results. This is useful to group a large amount of data and compute
# operation on these groups. Inside the agg() we specify to which column we want the
# calculation to be done over, and which type of calculations. We finish the exercise
# by printing the answer as an organized DataFrame.


answer4=data.groupby("Size").agg({"GDP per Capita":("mean","std")})
pd.DataFrame(answer4)
```

```python
#5.

# We start by defining the function "binary_v()", which takes a DataFrame and column
# as arguments. We are going to iterate over a list composed of the indexes of the
# DataFrame.

def binary_v(df,c_name):
    df1=df.copy()
    for code in df1.index.tolist():

# Then we create an if-else statement that will fill a new binary column according to
# to the values of the columns inserted as an argument. For each iteration over the column
# if the value of the selected column is NaN, the value in the new column will be NaN, if the
# value of the selected column is above the mean value of the column, the value in the new column
# will be 1, and finally, if the value of the selected column is below the mean value of the column,
# the value in the new column will be 0.

        if pd.isna(df1.loc[code,c_name]):
            df1.loc[code,c_name+" Binary"]=np.nan
        elif df1.loc[code,c_name]>df1[c_name].mean():
            df1.loc[code,c_name+" Binary"]=1
        else:
            df1.loc[code,c_name+" Binary"]=0
            
# Then the "bynary_v()" function will return the new column,
# without adding it to the initial defined DataFrame. That's why we copy the initial data frame to df1.
# We end the exercise by testing the new function and printing an example when using "GDP" as the column.
            
    return df1.loc[:,c_name+" Binary"]

pd.DataFrame(binary_v(data,"Population")).head(10)
```

```python
#6.

# Once again we are going to use the previously explained method .groupby(),
# but this time we are going to group by 2 variables, "Size" and "Human Capita Index Binary"
# (created with the previously defined function). Then we use the .agg() method to
# specify that we want to do the calculation over the "GDP per Capita" column, and the
# type of calculation we want to do is the average. We use method .unstack() as we have multiple indexes
# due to the fact we are grouping by two different columns.
# We end the exercise by printing a DataFrame as the answer to the question.

answer6=data.groupby(["Size",binary_v(data,"Human Capital Index")]).agg({"GDP per Capita":"mean"}).unstack()
pd.DataFrame(answer6)
```

```python
#7.

# This exercise follows the same reasoning as the previous one but here we want to
# group by "GINI Binary" and by "Human Capital Index Binary", we obtain both columns by
# using the function "binary_v()" created in question 5. We use method .unstack() due to 
# the reasons we previously explained. We finish the exercise by printing a DataFrame
# as the answer to the question.


answer7=data.groupby([binary_v(data,"Human Capital Index"),binary_v(data,"GINI")]).agg({"GDP per Capita":"mean"}).unstack()
pd.DataFrame(answer7)
```

```python
#8.

# We start the exercise by using the same reasoning and methods previously explained,
# the only difference is that instead of using the function mean in the .agg() method,
# we are using the function max, as we want the biggest GDP per capita for a below average
# Gini Index and a below average Human Capital Index.

df_8=data.groupby([binary_v(data,"GINI"),binary_v(data,"Human Capital Index")]).agg({"GDP per Capita":"max"}).unstack()

# Then, since we want the Gini Index and the Human Capital Index below average, we can slice
# the created DataFrame with .loc() and look directly for the GDP per Capita value we look for
# with that value we go to the original "data" DataFrame and we retrieve the respectively "Country Code".

code8=data[data["GDP per Capita"]==df_8.loc[0.0,('GDP per Capita', 0.0)]].index.tolist()[0]

# Then with the Country Code we go to the original "wdi" DataFrame to retrieve the name
# of the country and print it in the final answer as the question requests.

name8=wdi[["Country Name","Country Code"]].drop_duplicates()[wdi["Country Code"]==code8]["Country Name"].tolist()[0]
print(f"The country that has the highest GDP per capita + a Gini coefficient below average and a level of human capital below average is {name8}.")
```

```python
#9.

# In this exercise we will have to make a small change to the function defined in question 5.
# Now in the function our factor of decision between attributing NaN, 0 or 1, won't be the
# comparison between the column value and the mean of the column, but will be the comparison
# between the column value and the mean of that column taking into consideration the class size
# of the country we are looking at.

def binary_v_size(df,c_name):
    df1=df.copy()
    for code in data.index.tolist():
        if pd.isna(df.loc[code,c_name]):
            df1.loc[code,c_name+" Binary"]=np.nan
        elif df.loc[code,c_name]>df[df["Size"]==df.loc[code,"Size"]][c_name].mean():
            df1.loc[code,c_name+" Binary"]=1
        else:
            df1.loc[code,c_name+" Binary"]=0
    return df1.loc[:,c_name+" Binary"]


# Then we group by exactly using the same methods and reasoning as before.

df_9=data.groupby(["Size",binary_v_size(data,"GINI"),binary_v_size(data,"Human Capital Index")]).agg({"GDP per Capita":"max"}).unstack()

# Finally, we know that we are looking for the maximum GDP per Capita of a country
# with a Gini coefficient below average and a level of human capital index below average
# for its size classification. For this, we will have to iterate over a list containing 
# the different class sizes (we exclude NaN class size, that's why we index [:6]), and
# check which class has the highest GDP per capita taking into consideration the below
# average Human Capital Index and below average Gini Index within that class size.

list1=data["Size"].drop_duplicates().tolist()[:6]

max=0
for size in list1:
    if max<df_9.loc[(size, 0.0),('GDP per Capita', 0.0)]:
        max=df_9.loc[(size, 0.0),('GDP per Capita', 0.0)]
    else:
        pass
    
# To finish, now we know the maximum value for the GDP per Capita among the conditions
# we defined previously. With this value, we can look for the country code with the
# "data" DataFrame. With the country code and using the "wdi" DataFrame we can find
# the name of the country we are looking for, and finally print the final answer.

code9=data[data["GDP per Capita"]==max].index.tolist()[0]
name9=wdi[["Country Name","Country Code"]].drop_duplicates()[wdi["Country Code"]==code9]["Country Name"].tolist()[0]
print(f"The country that has the highest GDP per capita + a Gini coefficient below average for its size classification and a level of human capital below average for its size classification is {name9}.")
```
